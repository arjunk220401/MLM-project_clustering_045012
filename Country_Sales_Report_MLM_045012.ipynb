# 1 **Objectives**

1.**Clustering or segmenting the data**: Segmenting the data into clusters allows us to uncover underlying patterns and structures within the dataset, which can aid in understanding customer behavior, market trends, and product performance. However, it's essential to strike a balance between granularity and interpretability when determining the optimal number of clusters.

In this analysis, we employed several clustering algorithms (Mini-Batch K-Means, DBSCAN, Birch, and OPTICS) to segment the data based on features such as 'Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', and 'Total Profit'. Each algorithm generates a different number of clusters and provides insights into how the data can be grouped.

2.**No. of segments or clusters:** In this project, I have used 3 clusters with K =5,6 and 7.

These three clusters provide a good understanding of identifying patterns, similarities, and differences within the dataset. It facilitates the exploration of different customer segments, product categories, or market segments, enabling targeted strategies and decision-making.



3.**Characteristics of segments:**
With 3 clusters, the data is segmented into 3 distinct groups, each representing different patterns or behaviors within the dataset.

Cluster 1 may represent a group of observations with certain similarities or characteristics, distinct from those in Cluster 2 and Cluster 3.

Cluster 2 could capture another set of observations with different attributes compared to the other clusters.

Cluster 3 may indicate yet another subgroup within the dataset, potentially exhibiting unique behaviors or properties.

By examining the centroids or representative points of each cluster, it's possible to identify the key features or variables that distinguish one cluster from another.

Silhouette score provides a measure of how well-separated the clusters are, with higher values indicating better-defined clusters.
Davies-Bouldin score measures the average similarity between each cluster and its most similar one, with lower values indicating better clustering.

The training time and memory usage for 5 clusters are reasonable and comparable to those of 6 and 7 clusters.

Having fewer clusters (5 instead of 6 or 7) may lead to more interpretable results for managers, as it reduces complexity and makes it easier to understand and act upon the segmentation.

 While the differences in training time and memory usage between 5 and 6 or 7 clusters are minor, choosing 5 clusters still ensures optimal resource utilization without compromising performance.

Overall, choosing 5 clusters provides a good balance between model performance, interpretability, and resource efficiency, making it a suitable option for managers.
