```python
!pip install yellowbrick
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering, Birch
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
import time
import psutil
import matplotlib.pyplot as plt
from yellowbrick.cluster import KElbowVisualizer
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA

# Define a function to print memory usage
def get_memory_usage():
    process = psutil.Process()
    return process.memory_info().rss / 1024**2  # Memory usage in MB

# Read data from Excel file
df = pd.read_excel('100000 Sales Records.xlsx')
df.head()

# Drop unnecessary columns
columns_to_drop = ['Order Date', 'Order ID', 'Ship Date']
df = df.drop(columns=columns_to_drop)

# Handle missing values
imputer = SimpleImputer(strategy='mean')
df[['Unit Price', 'Unit Cost']] = imputer.fit_transform(df[['Unit Price', 'Unit Cost']])

# Encode categorical columns
label_encoder = LabelEncoder()
categorical_columns = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority']
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Handle remaining missing values (if any)
df = df.dropna()

# Standardize numerical features
numerical_columns = ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Split the data into features and labels
X = df.drop(columns=['Total Profit'])  # Features
y = df['Total Profit']  # Target variable

# Apply PCA to reduce dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Apply the elbow method to find the optimal number of clusters for K-Means
model_kmeans = KMeans(random_state=42)
visualizer_kmeans = KElbowVisualizer(model_kmeans, k=(3, 12), timings=False)
visualizer_kmeans.fit(X_pca)
visualizer_kmeans.show()

# Based on the elbow method, you can choose the optimal number of clusters (k) and update the code accordingly.
optimal_clusters = 4  # Update this based on the elbow method result

# Function to perform k-means clustering and measure time, memory, silhouette score, and Davies-Bouldin score
def perform_kmeans(X_data, num_clusters):
    start_time = time.time()

    # Create k-means object
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)

    # Fit and predict clusters
    clusters = kmeans.fit_predict(X_data)

    # Measure time taken
    time_taken = time.time() - start_time

    # Measure memory usage
    memory_usage = get_memory_usage()

    # Measure silhouette score
    silhouette = silhouette_score(X_data, clusters)

    # Measure Davies-Bouldin score
    db_score = davies_bouldin_score(X_data, clusters)

    return clusters, time_taken, memory_usage, silhouette, db_score

# Perform k-means clustering on the training set with the optimal number of clusters
train_clusters_optimal, train_time_optimal, train_memory_optimal, train_silhouette_optimal, train_db_score_optimal = perform_kmeans(X_train, num_clusters=optimal_clusters)

# Perform k-means clustering on the testing set with the optimal number of clusters
test_clusters_optimal, test_time_optimal, test_memory_optimal, test_silhouette_optimal, test_db_score_optimal = perform_kmeans(X_test, num_clusters=optimal_clusters)

# Display results for the optimal number of clusters for K-Means
print(f"K-Means Clustering (k={optimal_clusters}) - Training Set:")
print("Time Taken:", train_time_optimal, "seconds")
print("Memory Usage:", train_memory_optimal, "MB")
print("Silhouette Score:", train_silhouette_optimal)
print("Davies-Bouldin Score:", train_db_score_optimal)
print("Cluster Composition:")
print(pd.Series(train_clusters_optimal).value_counts())

print(f"\nK-Means Clustering (k={optimal_clusters}) - Testing Set:")
print("Time Taken:", test_time_optimal, "seconds")
print("Memory Usage:", test_memory_optimal, "MB")
print("Silhouette Score:", test_silhouette_optimal)
print("Davies-Bouldin Score:", test_db_score_optimal)
print("Cluster Composition:")
print(pd.Series(test_clusters_optimal).value_counts())

# Plot the clusters for the optimal number of clusters for K-Means
plt.scatter(X_test[:, 0], X_test[:, 1], c=test_clusters_optimal, cmap='viridis')
plt.title(f'K-Means Clustering (k={optimal_clusters}) - Testing Set')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Function to perform hierarchical clustering
def perform_hierarchical(X_data, num_clusters):
    start_time = time.time()

    # Create hierarchical clustering object
    hierarchical = AgglomerativeClustering(n_clusters=num_clusters)

    # Fit and predict clusters
    clusters = hierarchical.fit_predict(X_data)

    # Measure time taken
    time_taken = time.time() - start_time

    # Measure memory usage
    memory_usage = get_memory_usage()

    # Measure silhouette score (hierarchical clustering doesn't have silhouette score)
    silhouette = None

    # Measure Davies-Bouldin score
    db_score = davies_bouldin_score(X_data, clusters)

    return clusters, time_taken, memory_usage, silhouette, db_score

# Perform hierarchical clustering on the training set with the optimal number of clusters
train_clusters_hierarchical, train_time_hierarchical, train_memory_hierarchical, train_silhouette_hierarchical, train_db_score_hierarchical = perform_hierarchical(X_train, num_clusters=optimal_clusters)

# Perform hierarchical clustering on the testing set with the optimal number of clusters
test_clusters_hierarchical, test_time_hierarchical, test_memory_hierarchical, test_silhouette_hierarchical, test_db_score_hierarchical = perform_hierarchical(X_test, num_clusters=optimal_clusters)

# Display results for hierarchical clustering
print(f"Hierarchical Clustering (k={optimal_clusters}) - Training Set:")
print("Time Taken:", train_time_hierarchical, "seconds")
print("Memory Usage:", train_memory_hierarchical, "MB")
print("Davies-Bouldin Score:", train_db_score_hierarchical)
print("Cluster Composition:")
print(pd.Series(train_clusters_hierarchical).value_counts())

print(f"\nHierarchical Clustering (k={optimal_clusters}) - Testing Set:")
print("Time Taken:", test_time_hierarchical, "seconds")
print("Memory Usage:", test_memory_hierarchical, "MB")
print("Davies-Bouldin Score:", test_db_score_hierarchical)
print("Cluster Composition:")
print(pd.Series(test_clusters_hierarchical).value_counts())

```
